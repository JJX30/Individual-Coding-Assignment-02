# Individual Assignment 2

```python
import pandas as pd

# Read in the CSV file into the dataframe
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# Print out only the first 10 lines
print(train.head(10))
print(test.head(10))

# Print the number of missing values
print(train.isna().sum())
print(test.isna().sum())

# Replace empty cells with null
train.fillna("null", inplace=True)
test.fillna("null", inplace=True)

# Get random sample of n length
train = train.sample(n = 4000, random_state=90713366)
```

# Text preprocessing

```python
import re

# Create empty array for the first step
step_1_corpus = []

# Add to corpus array as lowercased
for sentence in train["Text"]:
    print(sentence)
    step_1_corpus.append(sentence.lower())

# Create empty array for the second step
step_2_corpus = []

# Remove the digits in the sentences 
for sentence in step_1_corpus:
    sentence = re.sub("[0-9]", "", sentence)
    step_2_corpus.append(sentence)

# Create empty array for the third step
step_3_corpus = []

# Fix contractions to be seperate words 
for sentence in step_2_corpus:
    sentence = re.sub(r"\'t", " not", sentence)
    sentence = re.sub(r"\'ve", " have", sentence)
    sentence = re.sub(r"\'s", " is", sentence)
    sentence = re.sub(r"\'re", " are", sentence)
    sentence = re.sub(r"can\'t", " can not", sentence)
    sentence = re.sub(r"\'ll", " will", sentence)
    sentence = re.sub(r"\'m", " am", sentence)
    sentence = re.sub(r"\'d", " would", sentence)
    sentence = re.sub(r"won\'t", " will not", sentence)

    step_3_corpus.append(sentence)


# Create empty array for the fourth step
step_4_corpus = []

for sentence in step_3_corpus:
    sentence = re.sub("[^a-z<> ]", "", sentence)
    step_4_corpus.append(sentence)

corpus = step_4_corpus
```

# Linguistic Feature Extraction

## Bag of words

```python
import nltk
import numpy
from nltk.corpus import stopwords

#try:
#    _create_unverified_https_context = ssl._create_unverified_context
#except AttributeError:
#    pass
#else:
#    ssl._create_default_https_context = _create_unverified_https_context
#nltk.download()

# Set the stopwords to english stop words
stop_words = set(stopwords.words("english"))

# Create empty array
new_corpus = []

# Make the sentences that get appended to the new array without the stop words
for sentence in corpus:
    sentence = [w for w  in sentence.split() if not w in stopwords.words("english")]
    new_corpus.append(sentence)

corpus_stop = []

for words in new_corpus:
    sentence = " ".join(words)
    corpus_stop.append(sentence)
# Print new corpus with no stopwords
print(corpus_stop)

word2count = {}

for sentence in corpus_stop:
    # This tokenizes the sentences, which means it gets the words as the elements
    words = nltk.word_tokenize(sentence)

    # Loop through the words in the array and also adding them to dictionary to count how many times it appears
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1

# Create list of unique words 
unique_words = list(word2count.keys())

# Create empty array of bag of words
bag_of_words = []

for sentence in corpus_stop:

    # Tokenizes each word in the sentence it is passed into
    words = nltk.word_tokenize(sentence)

    # This function creates an array with zeros with the length the size of the unique words array
    bag_vector = numpy.zeros(len(unique_words))

    for w in words:
        for i, word in enumerate(unique_words):
            if word == w:
                bag_vector[i] += 1

    # Finally appends words in the bag_vector into bag of words
    bag_of_words.append(bag_vector)
print(bag_of_words)
```
